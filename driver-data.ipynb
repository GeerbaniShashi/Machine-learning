```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (12,6)
```

```python
df = pd.read_csv('/content/gdrive/My Drive/At a glance: Critical Care Unit/driver-data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2)
df_analyze = df.drop('id', axis = 1)
```

```python
print(kmeans.fit(df_analyze))
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
print(len(kmeans.labels_))
```

```python
print(type(kmeans.labels_))
unique,counts = np.unique(kmeans.labels_, return_counts = True)
print(dict(zip(unique,counts)))
```

```python
df_analyze['cluster'] = kmeans.labels_
sns.set_style('whitegrid')
sns.lmplot('mean_dist_day', 'mean_over_speed_perc', data=df_analyze,
           hue='cluster', palette='coolwarm', height=6, aspect=1, fit_reg=False)
```
