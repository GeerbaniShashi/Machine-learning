```python
#array manipulation - changing shape
```

```python
import numpy as np
a = np.arange(9)
print('The original array: ')
print(a)
print()

b= a.reshape(3,3)
print('The modified array: ')
print(b)
```

```python
print(b.flatten())
```

```python
b.flatten(order="F")
```

```python
b.flatten(order="C")
```

```python
b.flatten(order="A")
```

```python
a = np.arange(12).reshape(4,3)
a
```

```python
print(np.transpose(a))
```

```python
b = np.arange(8).reshape(2,4)
b
```

```python
c = b.reshape(2,2,2)
c
```

```python
np.rollaxis(c,1,2)
```

```python
np.swapaxes(c,1,2)
```
