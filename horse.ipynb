```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
animals = pd.read_csv("/content/gdrive/My Drive/At a glance: Critical Care Unit/horse1.csv")
animals
```

```python
target = animals["outcome"]
```

```python
target.unique()
```

```python
animals = animals.drop(['outcome'],axis=1)
```

```python
category_variables = ['surgery','age','temp_of_extremities','peripheral_pulse','mucous_membrane','capillary_refill_time','pain','peristalsis','abdominal_distention','nasogastric_tube','nasogastric_reflux','rectal_exam_faces','abdomen','abdomo_appearance','surgical_lesion','cp_data']
for category in category_variables:
   animals[category] = pd.get_dummies(animals[category])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
X,y = animals.values,target.values
label_encoder = LabelEncoder()
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier
print(X_train.shape)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imp = SimpleImputer(missing_values = np.nan,strategy = "most_frequent")
X_train = imp.fit_transform(X_train)
X_test = imp.fit_transform(X_test)
```

```python
classifier = DecisionTreeClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_predict,y_test)
```

```python
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
```

```python
classifier.fit(X_train,y_train)
y_predict = classifier.predict(X_test)
accuracy = accuracy_score(y_predict,y_test)
print(accuracy)
```
