```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,4,8,16]

plt.plot(x,y1,label = 'Line 1')
plt.plot(x,y2,label = 'Line 2')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['A','B','C','D','E']
y1 = [10,24,36,21,14]
y2 = [5,12,18,10,7]

plt.bar(x,y1,label = 'Group 1')
plt.bar(x,y2,bottom = y1, label = 'Group 2')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X,Y = np.meshgrid(x,y)
Z = np.exp(-X**2 -Y**2)

plt.contour(X,Y,Z)
plt.show()
```
